### **A)**

### $\frac{V_B-V}{R_3} = \frac {V}{R_4}$  
### $V_B = (\frac 1{R_4}+\frac 1{R_3})V{R_3}$      

---------------

### $I_{C1} = I_{R2}$  
### $\frac{V-V_A}{\frac1{SC_1}} = \frac {V_B-V}{R_2}$  
### $(SC_1+\frac1{R_2})V = SC_1V_A+\frac {V_B}{R_2}$      
### $(SC_1+\frac1{R_2})V = SC_1V_A+\frac {R_3+R_4}{R_2R_4}V$      
### $(SC_1+\frac1{R_2}-\frac {R_3+R_4}{R_2R_4})V = SC_1V_A$      
### $(SC_1+\frac {R_4-R_3-R_4}{R_2R_4})V = SC_1V_A$      
### $(1-\frac {R_3}{R_2R_4C_1S})V = V_A$      

---------------

### $I=I_{R1}=\frac V{R_1}-\frac 1{R_1}\frac {R_2R_4C_1S-R_3}{R_2R_4C_1S}V=\frac {R_2R_4C_1S+R_3-R_2R_4C_1S}{R_1R_2R_4C_1S}V$ 
### $\frac VI=\boxed {Z_e=\frac {R_1R_2R_4C_1}{R_3}S}$  
## Es una inductancia activa de $L=\frac{R_1R_2R_4C_1}{R_3}$  
---------------

### ${Z_e(S)}_{S=j\omega}=j\omega L$  
### $\boxed{\left|{Z_e(j\omega)}\right|=\omega L}$  
### $\phi [Z_e(j\omega)]=\arctan(\frac {\omega L}0)=\arctan(\rightarrow \infty)$  
### $\boxed{\phi [Z_e(j\omega)]=\frac \pi2}$

### $\frac {I_1}{gm_2}=\frac{V_igm_1}{SC_1}$

### **C)**

### $I_{R3} = I_{R4}$  
### $\frac{V_B-V}{R_3} = \frac {V}{R_4}$  
### $V_B = (\frac {R_3}{R_4}+1)V$      

---------------

### $I_{R1} = I_{C2}$  
### $\frac{V-V_A}{R_1} = (V_B-V)C_2S$  
### $(SC_2+\frac1{R_1})V = \frac{V_A}{R_1}+(\frac {R_3}{R_4}+1)C_2SV$  
### $(\frac1{R_1}+SC_2-\frac {R_3+R_4}{R_4}C_2S)V = \frac{V_A}{R_1}$      
### $(\frac1{R_1}-\frac {R_3}{R_4}C_2S)V = \frac{V_A}{R_1}$   
### $(1-\frac {R_1R_3C_2}{R_4}S)V = V_A$      

---------------

### $I_{C1}=I=(V-V_A)C_1S$  
### $I=(V-(1-\frac {R_1R_3C_2}{R_4}S)V)C_1S$  
### $I=(1-1+\frac {R_1R_3C_2}{R_4})C_1SV$  
### $\frac VI=\boxed {Z_e=\frac {R_4}{R_1R_3C_1C_2}\frac1{S^2}}$  
## Es un integrador de orden 2   
---------------
### ${Z_e(S)}=10^9\frac1{S^2}$

### $\boxed{\left|{Z_e(j\omega)}\right|=10^9\frac1{\omega^2}}$   
### $\boxed{\phi [Z_e(j\omega)]=\pi}$
