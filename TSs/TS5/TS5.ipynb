### Normalización de plantilla pasa banda objetivo:

### $\omega_0= 2\,\pi\,22000 \frac{rad}{seg}$  

### $\omega_{S1}= 2\,\pi\,17000 \frac{rad}{seg}$  

### $\omega_{S2}= 2\,\pi\,36000 \frac{rad}{seg}$  

### Norma de normalización: $\Omega_\omega= \omega_0= 2\,\pi\,22000 \frac{rad}{seg}$  

### $\omega_{0_N}= \frac{\omega_0}{\Omega_\omega}= \frac{\omega_0}{\omega_0}= \frac{2\,\pi\,22000 [\frac{rad}{seg}]}{2\,\pi\,22000 [\frac{rad}{seg}]} \Rightarrow  \boxed{\omega_{0_N}=1}$  

### $\omega_{S1_N}= \frac{\omega_{S1}}{\Omega_\omega}= \frac{\omega_{S1}}{\omega_0}= \frac{2\,\pi\,17000 [\frac{rad}{seg}]}{2\,\pi\,22000 [\frac{rad}{seg}]} \Rightarrow  \boxed{\omega_{S1_N}=0,7727}$  

### $\omega_{S2_N}= \frac{\omega_{S2}}{\Omega_\omega}= \frac{\omega_{S2}}{\omega_0}= \frac{2\,\pi\,36000 [\frac{rad}{seg}]}{2\,\pi\,22000 [\frac{rad}{seg}]} \Rightarrow  \boxed{\omega_{S2_N}=1,6364}$  

### Se hallan valores de $\omega_{P1_N}\,$ y $\omega_{P2_N}$:  

### (1) $\frac{\omega_{0_N}}{Q}=B=\omega_{P2_N}-\omega_{P1_N}$  
### (2) ${\omega_{0_N}}^2=\omega_{P1_N}\,\omega_{P2_N}$  

### Resolviendo sistema de ecuaciones:

### $\rightarrow \omega_{P1_N} = -1,105\,$ se descarta pues $\omega_{P1_N}>0$  
### $\Rightarrow \boxed{\omega_{P1_N} = 0,905}\,\,\, \land \,\,\, \boxed{\omega_{P2_N} = 1,105}$  

### Y así queda definida la plantilla pasa banda objetivo normalizada (los valores de atenuación no se alteran)

<br>

---------------

### Se transforman las frecuencias normalizadas a pasa bajos prototipo aplicando núcleo de transformación:  

$K(\omega)=\Omega=Q \,\frac {\omega^2-{\omega_0}^2}\omega$  
### siendo:  
### $s=\sigma + j \omega \,$ (pasa banda objetivo)  
### $p=\Sigma+j \Omega \,$ (pasa bajos prototipo)  
<br>

### $\rightarrow\, \Omega_{P1_N}=Q \,\frac {{\omega_{P1_N}}^2-{\omega_0}^2}{\omega_{P1_N}}\,\Rightarrow\, \Omega_{P1_N}=-1$    
### $\rightarrow\, \Omega_{P2_N}=Q \,\frac {{\omega_{P2_N}}^2-{\omega_0}^2}{\omega_{P2_N}}\,\Rightarrow\, \Omega_{P2_N}=1$    
### $\rightarrow\, \Omega_{S1_N}=Q \,\frac {{\omega_{S1_N}}^2-{\omega_0}^2}{\omega_{S1_N}}\,\Rightarrow\, \Omega_{S1_N}=-2,6073$  
### $\rightarrow\, \Omega_{S2_N}=Q \,\frac {{\omega_{S2_N}}^2-{\omega_0}^2}{\omega_{S2_N}}\,\Rightarrow\, \Omega_{S2_N}=5,1265$

<br>

---------------

### Se obtiene el $\,\epsilon\,$ y también el orden del filtro Chebyshev pasabajos prototipo para que cumpla el requerimiento especificado en plantilla en $\,\Omega_{S1_N}\,$ y $\,\Omega_{S2_N}\,\,\, \rightarrow\,$ se elige el orden mayor de ellas (roll-off más exigente):  

### $\epsilon^2 = 10^{\frac{\alpha_{máx}}{10}}-1 = 10^{0,05}-1 \approx 0,1222 \,\Rightarrow\, \epsilon \approx 0,3493$  

### A) Se analiza el caso de $\Omega_{S1_N}$:  

- ### $\Omega_P=|\Omega_{P1_N}|=1$
- ### $\Omega_S=|\Omega_{S1_N}|=2,6073$
- ### $\alpha_{máx}=0,5\,dB$  
- ### $\alpha_{mín}=16\,dB$
<br>

### $α_{dB}= 10\,log[1+ε^2\,C_n^2(Ω)]$  

### $(\alpha_{mín}, \Omega_S): \alpha_{mín} = 10\,log[1 + ε^2\,C_n^2(\Omega_S)]$

### $\,n_1= 1: ¿\alpha(\Omega_S) = 2,62 dB \geq \alpha_{mín}= 16 dB? \rightarrow No...$  

### $\,n_1= 2: ¿\alpha(\Omega_S) = 13,09 dB \geq \alpha_{mín}= 16 dB? \rightarrow No...$  

### $\,n_1= 3: ¿\alpha(\Omega_S) = 26,87 dB \geq \alpha_{mín}= 16 dB? \rightarrow Sí...$  

### $\Rightarrow n_1=3$

### B) Se analiza el caso de $\Omega_{S2_N}$:  

- ### $\Omega_P=\Omega_{P2_N}=1$
- ### $\Omega_S=\Omega_{S2_N}=5,1265$
- ### $\alpha_{máx}=0,5\,dB$  
- ### $\alpha_{mín}=24\,dB$
<br>

### $α_{dB}= 10\,log[1+ε^2\,C_n^2(Ω)]$  

### $(\alpha_{mín}, \Omega_S): \alpha_{mín} = 10\,log[1 + ε^2\,C_n^2(\Omega_S)]$

### $\,n_2= 1: ¿\alpha(\Omega_S) = 6,24 dB \geq \alpha_{mín}= 24 dB? \rightarrow No...$  

### $\,n_2= 2: ¿\alpha(\Omega_S) = 25,12 dB \geq \alpha_{mín}= 24 dB? \rightarrow Sí...$  

### $\Rightarrow n_2=2$
<br>

### Se elige el orden mayor $\,\Rightarrow n=3$

<br>

---------------

### Se halla transferencia pasa bajos prototipo normalizada:  

### $|T_{LP_N}(jΩ)|^2= \frac{1}{1+ε^2\,C_3^2(Ω)}=\frac{1}{1+ε^2\,(4\Omega^3-3\Omega)^2}=\frac{\frac 1 {16\epsilon^2}}{\Omega^6-\frac 3 2 \Omega^4+\frac 9 {16} \Omega^2+\frac 1 {16\epsilon^2}}$  
### ${{|T_{LP_N}(jΩ)|}^2}_{\Omega=\frac p j}=|T_{LP_N}(p)|^2 =\frac{\frac 1 {16\epsilon^2}}{-p^6-\frac 3 2 p^4-\frac 9 {16} p^2+\frac 1 {16\epsilon^2}}= T(p)\,T(-p) = \frac a {p^3+bp^2+cp+d}\,\frac a {-p^3+bp^2-cp+d}$  

### $\frac{\frac 1 {16\epsilon^2}}{-p^6-\frac 3 2 p^4-\frac 9 {16} p^2+\frac 1 {16\epsilon^2}}= \frac a {p^3+bp^2+cp+d}\,\frac a {-p^3+bp^2-cp+d}$  

### y quedan restringidos los valores de los coeficientes:  
### $a^2 = \frac 1 {16\epsilon^2} \, \Rightarrow \, a=\frac 1 {4\epsilon}=0,7157\,$ (me quedo con valor positivo para que no haya desfasaje de 180º en la transferencia)  
### $d^2 = \frac 1 {16\epsilon^2} \, \Rightarrow \, d=\frac 1 {4\epsilon}=0,7157\,$ (me quedo con valor positivo para que el sistema sea estable, polos en semiplano izquierdo)  
### $2 b d - c^2=-\frac 9 {16} \,\, \land \,\, b^2-2 c=-\frac 3 2 \,$ resolviendo sistema de ecuaciones cuadrático $\, \rightarrow \, b=1,2529 \land \,c=1,5349$  
<br>

### Finalmente queda definida la transferencia pasabajos prototipo normalizada:  
### $\boxed{T_{{LP}_N}(p)=\frac {0,7157}{p^3+ 1,2529\,p^2 +1,5349\,p + 0,7157}=\frac {0,6265}{p + 0,6265}\,\,\,\frac {1,1424}{p^2 + 0,6265\,p + 1,1424}}$

## **Simulación numérica de función transferencia pasa bajos prototipo normalizada con Python**

Cargo librerías:

```python
# módulos numéricos y de funciones científicas
import numpy as np
from scipy import signal as sig
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import matplotlib as mpl

# PyTC2: La librería para TC2
from pytc2.sistemas_lineales import bodePlot, pzmap, GroupDelay, analyze_sys, pretty_print_lti
```

Configuro gráficos:

```python
plt.figure(1)
plt.close(1)

fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80
fig_font_size = 16

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
plt.rcParams.update({'font.size':fig_font_size})
```

Simulo:

```python
#-------------------------------------------------------------------------------------------------------------------#

# Datos de plantilla

alfa_max = 0.5                        # dB
alfa_min = 16                       # dB

f_0_plantilla = 22e3               # Hz
f_s1_plantilla = 17e3               # Hz
f_s2_plantilla = 36e3               # Hz

omega_0_plantilla = 2*np.pi*f_0_plantilla
omega_s1_plantilla = 2*np.pi*f_s1_plantilla
omega_s2_plantilla = 2*np.pi*f_s2_plantilla

norm_w = omega_0_plantilla

omega_0_norm = omega_0_plantilla/norm_w
omega_s1_norm = omega_s1_plantilla/norm_w
omega_s2_norm = omega_s2_plantilla/norm_w
omega_p1_norm = 0.905               # Hz
omega_p2_norm = 1.105               # Hz


print('#---------------------------------------------------------------------------------------------#')
print(' ')
print('-- Datos de plantilla - Filtro pasa banda objetivo --')
print(' ')

print('α_min = {}  (se utiliza la más exigente)'.format(alfa_min))
print('α_max = {}'.format(alfa_max))
print(' ')
print('f0 = {}'.format(f_0_plantilla))
print('fs1 = {}'.format(f_s1_plantilla))
print('fs2 = {}'.format(f_s2_plantilla))
print(' ')
print('ω_0_N = {:3.4f}'.format(omega_0_norm))
print('ω_p1_N = {:3.4f}'.format(omega_p1_norm))
print('ω_p2_N = {:3.4f}'.format(omega_p2_norm))
print('ω_s1_N = {:3.4f}'.format(omega_s1_norm))
print('ω_s2_N = {:3.4f}'.format(omega_s2_norm))


#-------------------------------------------------------------------------------------------------------------------#

# Obtenemos orden y omega_0 del filtro

n, wn = sig.cheb1ord([omega_p1_norm,omega_p2_norm], [omega_s1_norm,omega_s2_norm], alfa_max, alfa_min, analog=True)

print(' ')
print('#---------------------------------------------------------------------------------------------#')
print(' ')
print('Orden del filtro: n = {}'.format(n))
print('Frecuencias angulares de paso: ω_p_N_1;2 = {}'.format(wn))

#-------------------------------------------------------------------------------------------------------------------#

# Obtenemos transferencia pasabajos prototipo normalizada

z, p, k = sig.cheb1ap(n, alfa_max)

print(' ')
print(' ')
print('#---------------------------------------------------------------------------------------------#')
print('Función transferencia pasa bajos prototipo normalizada:')
print(' ')

print('-- z: Ceros coincidentes con los obtenidos de forma analítica --')
print(' ')

print('z= {}'.format(z))

print(' ')
print('-- p: Polos coincidentes con los obtenidos de forma analítica --')
print(' ')

print('p= {}'.format(p))

print(' ')
print('-- k: Ganancia coincidentes con la obtenida de forma analítica --')
print(' ')

print('k= {}'.format(k))

# Transformamos las singularidades que nos devuelve cheb1ap() para obtener 
# el numerador y el denominador de la T_LP_N(s) del filtro pasa bajos prototipo normalizado:

num_lp, den_lp = sig.zpk2tf(z,p,k)

# Imprimimos la T_LP_N(s) en formato T(s)= num/den con la funcion pretty_print_lti(num, den):

print(' ')
print(' ')
print('#---------------------------------------------------------------------------------------------#')
print(' ')
print('T_LP_N(s):')
pretty_print_lti(num_lp, den_lp)
print(' ')
print('#---------------------------------------------------------------------------------------------#')
print(' ')

TS5 = sig.TransferFunction(num_lp,den_lp)

analyze_sys(TS5, 'T_LP_N')
```

## **Simulación numérica de función transferencia pasa banda objetivo normalizada con Python**

```python
#-------------------------------------------------------------------------------------------------------------------#

# Obtenemos la transferencia pasa banda objetivo directamente con las especificiones de plantilla cargadas anteriormente

z,p,k = sig.cheby1(n, alfa_max, wn , btype='bandpass', analog=True, output='zpk')

print(' ')
print('#---------------------------------------------------------------------------------------------#')

print(' Filtro pasa banda objetivo')
print('#---------------------------------------------------------------------------------------------#')

print('-- z: Ceros coincidentes con los obtenidos de forma analítica --')
print(' ')

print('z= {}'.format(z))

print(' ')
print('-- p: Polos coincidentes con los obtenidos de forma analítica --')
print(' ')

print('p= {}'.format(p))

print(' ')
print('-- k: Ganancia obtenida de forma analítica --')
print(' ')

print('k= {}'.format(k))

# Transformamos las singularidades que nos devuelve cheb1ap() para obtener 
# el numerador y el denominador de la T_BP_N(s) del filtro pasa banda objetivo normalizado:

num_bp, den_bp = sig.zpk2tf(z,p,k)

# Imprimimos la T_BP_N(s) en formato T(s)= num/den con la funcion pretty_print_lti(num, den):

print(' ')
print(' ')
print('#---------------------------------------------------------------------------------------------#')
print(' ')
print('T_BP_N(s):')
pretty_print_lti(num_bp, den_bp)
print(' ')
print('#---------------------------------------------------------------------------------------------#')
print(' ')

TS5 = sig.TransferFunction(num_bp,den_bp)

analyze_sys(TS5, 'T_BP_N')
```
