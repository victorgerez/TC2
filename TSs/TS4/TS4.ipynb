```python
# m√≥dulos num√©ricos y de funciones cient√≠ficas
import numpy as np
from scipy import signal as sig
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import matplotlib as mpl

# PyTC2: La librer√≠a para TC2
from pytc2.sistemas_lineales import bodePlot, pzmap, GroupDelay, analyze_sys
```

### Factorizo la transferencia dividi√©ndola por el binomio $s+\epsilon^{-\frac 1 3}$ (ra√≠z hallada) mediante regla de Ruffini:  
### $(s^3+2 \epsilon^{-\frac 1 3} s^2 + 2 \epsilon^{-\frac 2 3} s + \frac 1 \epsilon) \div (s+\epsilon^{-\frac 1 3}) = s^2+ \epsilon^{-\frac 1 3} s + \epsilon^{-\frac 2 3}$

### Respuesta en frecuencia:  
### $T(s)|_{s=j\omega}=T(j\omega)=\frac {\frac 1 \epsilon}{-j\omega^3-2 \epsilon^{-\frac 1 3} \omega^2 + 2 \epsilon^{-\frac 2 3} j\omega + \frac 1 \epsilon} = \frac {\frac 1 \epsilon}{-2 \epsilon^{-\frac 1 3} \omega^2  + \frac 1 \epsilon +j(-\omega^3 + 2 \epsilon^{-\frac 2 3} \omega)}$ 
### $\boxed{\left|{T(j\omega)}\right|= \frac 1{\sqrt{1+\epsilon^2 \omega^6}}}$
### Si $\omega = 0 \, \Rightarrow \, |{T(j\omega)}|= 1 = 0 dB$

### Si $\omega \rightarrow \infty \, \Rightarrow \, |{T(j\omega)}|= 0$
<br>

### $\phi [T(j\omega)]=\phi[N(j\omega)]-\phi[D(j\omega)]=\arctan(\frac {\Im[N(j\omega)]}{\Re[N(j\omega)]})-\arctan(\frac {\Im[D(j\omega)]}{\Re[D(j\omega)]})=0-\arctan(\frac {-\omega^3 + 2 \epsilon^{-\frac 2 3} \omega)}{-2 \epsilon^{-\frac 1 3} \omega^2  + \frac 1 \epsilon})$  
### Y como $\, \arctan(x)=-\arctan(-x)\,$ por ser funci√≥n impar:  
### $\boxed{\phi [T(j\omega)]=\arctan(\frac {\omega^3 - 2 \epsilon^{-\frac 2 3} \omega)}{2 \epsilon^{-\frac 1 3} \omega^2  - \frac 1 \epsilon})}$  
### Si $\omega = 0 \, \Rightarrow \, \phi [T(j\omega)]= 0 $  
### Si $\omega = \epsilon^{-\frac 1 3} \, \Rightarrow \, \phi [T(j\omega)]= -\frac 3 4 \pi $

### Para implementar esta transferencia pasabajos de orden 3 con estructuras pasivas cascadeo una estructura de orden 1 (R-C serie) con otra de orden 2 (R-L-C serie), ambas pasabajos.

## $T(s)=\frac {\frac 1 \epsilon}{s^3+2 \epsilon^{-\frac 1 3} s^2 + 2 \epsilon^{-\frac 2 3} s + \frac 1 \epsilon}= \frac {\epsilon^{-\frac 1 3}}{s+\epsilon^{-\frac 1 3}} \,\, \frac{\epsilon^{-\frac 2 3}}{s^2+ \epsilon^{-\frac 1 3} s + \epsilon^{-\frac 2 3}}$
<br>

### Orden 1 (R-C serie):  
## $T_1(s)= \frac {\frac 1{R_1 C_1}} {s+\frac 1 {R_1 C_1}}=\frac {\epsilon^{-\frac 1 3}}{s+\epsilon^{-\frac 1 3}}$

<br>

### Orden 2 (R-L-C serie):  
## $T_2(s)= \frac {K\omega_0^2}{s^2+\frac {\omega_0}Q s+\omega_0^2}= \frac {K {\sqrt {\frac 1 {L_2 C_2}}}^2}{s^2+\frac {R_2} {L_2} s+\sqrt {\frac 1 {L_2 C_2}}^2} =\frac{\epsilon^{-\frac 2 3}}{s^2+ \epsilon^{-\frac 1 3} s + \epsilon^{-\frac 2 3}}$  
### $\Rightarrow \, K=1 \, \land \,\omega_0=\epsilon^{-\frac 1 3} \, \land \,Q=1\,\,$ y los componentes:
### $\sqrt {\frac 1 {L_2 C_2}}=\omega_0=\epsilon^{-\frac 1 3}\,\rightarrow \,$ adopto: $\boxed{C_{2_N=1}} \, \therefore \, \boxed{L_{2_N}=\epsilon^{\frac 2 3}}$    
### $\frac {R_2}{L_2}=\frac {\omega_0} Q = \epsilon^{-\frac 1 3} \, \Rightarrow \boxed{R_{2_N}=\epsilon^{\frac 1 3}}$
### y de esta forma $C_{1_N} = C_{2_N} = C_N = 1\,\,$ ; $\,\,R_{1_N} = R_{2_N} = R_N = \epsilon^{\frac 1 3}\,\,$ y $\,\, L_{2_N}=L_N=R_N^2=\epsilon^{\frac 2 3}\,\,$ (recordar que todos estos valores son normalizados)
