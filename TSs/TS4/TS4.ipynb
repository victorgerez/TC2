## **Simulación simbólica de funciones transferencia con Python**

```python
#-------------------------------------------------------------------------------------------------------------------#

# Datos de plantilla

alfa_max = 1                        # dB
alfa_min = 30                       # dB

f_s_plantilla = 10e3               # Hz
f_p_plantilla = 40e3               # Hz

omega_p_plantilla = 2*np.pi*f_p_plantilla
omega_s_plantilla = 2*np.pi*f_s_plantilla

norm_w = omega_p_plantilla

omega_p_norm = omega_p_plantilla/norm_w
omega_s_norm = omega_s_plantilla/norm_w

omega_p_pb_prototipo = 1/omega_p_norm
omega_s_pb_prototipo = 1/omega_s_norm

print('#---------------------------------------------------------------------------------------------#')
print(' ')
print('-- Datos de plantilla - Filtro pasa altos objetivo --')
print(' ')

print('α_min = {}'.format(alfa_min))
print('α_max = {}'.format(alfa_max))
print('fs = {}'.format(f_s_plantilla))
print('fp = {}'.format(f_p_plantilla))

print(' ')

print('ωp = {:3.4f}'.format(omega_p_plantilla))
print('ωs = {:3.4f}'.format(omega_s_plantilla))
print('ωp_n = {:3.4f}'.format(omega_p_norm))
print('ωs_n = {:3.4f}'.format(omega_s_norm))

print(' ')
print('-- Dominio transformado - Filtro pasa bajos prototipo --')
print(' ')

print('Ωp = {:3.4f}'.format(omega_p_pb_prototipo))
print('Ωs = {:3.4f}'.format(omega_s_pb_prototipo))

#-------------------------------------------------------------------------------------------------------------------#

# Obtenemos orden y omega 0 del filtro pasa bajos prototipo

n, w0 = sig.buttord(omega_p_pb_prototipo, omega_s_pb_prototipo, alfa_max, alfa_min, analog=True)


print('#---------------------------------------------------------------------------------------------#')
print(' ')
print('Orden del filtro: n = {}'.format(n))
print('Frecuencia angular de corte: w0 = {}'.format(w0))

#-------------------------------------------------------------------------------------------------------------------#

# Obtenemos filtro pasa bajos prototipo de máxima planicidad

z,p,k = sig.butter(n, w0, btype='lowpass', analog=True, output='zpk')

print(' ')
print('#---------------------------------------------------------------------------------------------#')

print(' Filtro pasa bajos prototipo')
print('#---------------------------------------------------------------------------------------------#')

print('-- z: Ceros coincidentes con los obtenidos de forma analítica --')
print(' ')

print('z= {}'.format(z))

print(' ')
print('-- p: Polos coincidentes con los obtenidos de forma analítica --')
print(' ')

print('p= {}'.format(p))

# Transformamos las singularidades que nos devuelve butter() para obtener 
# el numerador y el denominador de la T_LP_N(s) del filtro pasa bajos prototipo normalizado:

num_lp, den_lp = sig.zpk2tf(z,p,k)

# Imprimimos la T_LP_N(s) en formato T(s)= num/den con la funcion pretty_print_lti(num, den):

print(' ')
print('#---------------------------------------------------------------------------------------------#')
print('-- Función transferencia T_LP(s) normalizada:')
print(' ')

pretty_print_lti(num_lp, den_lp)

#-------------------------------------------------------------------------------------------------------------------#

# Transformamos la T_LP_N(s) del filtro pasa bajos prototipo para obtener 
# el numerador y el denominador de la T_HP_N(s) con lp2hp(num, den):
print(' ')
print('#---------------------------------------------------------------------------------------------#')

print(' Filtro pasa altos objetivo')
print('#---------------------------------------------------------------------------------------------#')

num_hp, den_hp = sig.lp2hp(num_lp, den_lp)

print(' ')
print('-- z: Ceros coincidentes con los obtenidos de forma analítica --')
print(' ')

print('z1,2= {}'.format(np.roots(num_hp)))

print(' ')
print('-- p: Polos coincidentes con los obtenidos de forma analítica --')
print(' ')

print('p1,2= {}'.format(np.roots(den_hp)))

print(' ')
print('#---------------------------------------------------------------------------------------------#')

# Imprimimos la T_HP_N(s) en formato T(s)= num/den con la funcion pretty_print_lti(num, den):

print(' ')
print('#---------------------------------------------------------------------------------------------#')
print('-- Función transferencia T_HP(s) normalizada:')
print(' ')

pretty_print_lti(num_hp, den_hp)

#-------------------------------------------------------------------------------------------------------------------#


T_HP_N= sig.TransferFunction(num_hp, den_hp)                   # numerador y denominador con butter()

#-------------------------------------------------------------------------------------------------------------------#

# Representamos módulo, fase, diagrama de polos y ceros y retardo de fase con la analyze_sys() que tiene, entre
# otras cosas, indicadores (w0 y Q) de las singularidades en el diagrama de polos y ceros

print(' ')
print('#---------------------------------------------------------------------------------------------#')
print('-- Módulo, fase, diagrama de polos y ceros y retardo de fase --')
print(' ')

analyze_sys(T_HP_N, sys_name='T_HP_N')

#-------------------------------------------------------------------------------------------------------------------#
```

### Para comparar este diagrama de polos y ceros con el de la transferencia pasa bajos prototipo normalizada, se simulará con Python:  

```python
epsilon = 0.509
erc = 1/np.cbrt(epsilon) # epsilon^(-1/3)

num = np.array([0,0,0,erc*erc*erc]) 
den = np.array([1,2*erc,2*erc*erc,erc*erc*erc])

TS = sig.TransferFunction(num,den)

analyze_sys(TS, 'T_LP_N')
```

<br>

### Se puede notar que:

### - Ceros: mientras que en el pasa bajos prototipo están en el infinito (como todo pasa bajos) en el pasa altos objetivo están en el origen (como todo pasa altos), se aprecia la naturaleza inversa de ambos filtros, que matemáticamente se visualiza en $K(s)=p=\frac 1 s$  

### - Polos: son los mismos, por ende ambas transferencias tienen el mismo $\,Q\,$ y $\,\omega_0\,$, es decir mismo roll-off en banda de transición, mismo ripple de máxima planicidad en banda de paso y todo esto sobre la mismas frecuencias (no olvidar que están normalizadas).
<br>

------------

### Respuesta en frecuencia:  
## ${T_{{HP}_N}(s)|_{s=j\omega}=T_{{HP}_N}(j\omega)=\frac {-j\omega^3}{-j\omega^3-2 \epsilon^{\frac 1 3} \omega^2 + 2 \epsilon^{\frac 2 3} j \omega + \epsilon}} = \frac {-j\omega^3}{-2 \epsilon^{\frac 1 3} \omega^2 + \epsilon +j(-\omega^3 + 2 \epsilon^{\frac 2 3} \omega)}$ 
### $\boxed{|T_{{HP}_N}(j\omega)|= \frac {\omega^3}{\sqrt{(-2 \epsilon^{\frac 1 3} \omega^2 + \epsilon)^2 +(-\omega^3 + 2 \epsilon^{\frac 2 3} \omega)^2}}}$
### Si $\omega = 0 \, \Rightarrow \, |T_{{HP}_N}(j0)| = 0$
### Si $\omega = \epsilon^{\frac 1 3} \, \Rightarrow \, |{T(j\epsilon^{\frac 1 3})}|= \frac 1 {\sqrt 2} = -3 dB$
### Si $\omega \rightarrow \infty \, \Rightarrow \, |{T(\rightarrow \infty)}|= 1 = 0 dB$
<br>

### $\phi [T_{{HP}_N}(j\omega)]=\phi[N(j\omega)]-\phi[D(j\omega)]=\arctan(\frac {\Im[N(j\omega)]}{\Re[N(j\omega)]})-\arctan(\frac {\Im[D(j\omega)]}{\Re[D(j\omega)]})=-\frac {\pi}2-\arctan(\frac {-\omega^3 + 2 \epsilon^{\frac 2 3} \omega)}{-2 \epsilon^{\frac 1 3} \omega^2  + \epsilon})$  
### Y como $\, \arctan(x)=-\arctan(-x)\,$ por ser función impar:  
### $\boxed{\phi [T_{{HP}_N}(j\omega))]=-\frac {\pi}2+\arctan(\frac {\omega^3 - 2 \epsilon^{\frac 2 3} \omega)}{2 \epsilon^{\frac 1 3} \omega^2  - \epsilon})}$  
### Si $\omega = 0 \, \Rightarrow \, \phi [T_{{HP}_N}(j0)]= -\frac \pi 2 $  
### Si $\omega = \epsilon^{\frac 1 3} \, \Rightarrow \, \phi [T_{{HP}_N}(j\epsilon^{\frac 1 3})]= -\frac 5 4 \pi $  
### Si $\omega \rightarrow \infty \, \Rightarrow \, \phi [T_{{HP}_N}(\rightarrow \infty)]= - 2 \pi$

### Para implementar esta transferencia pasa altos normalizada de orden 3 con estructuras pasivas cascadeo una estructura de orden 1 (C-R serie) con otra de orden 2 (R-C-L serie), ambas pasa altos.

## $T_{{HP}_N}(s)=\frac s{s+\omega_0} \, \frac {s^2}{s^2+\frac {\omega_0}Q s+\omega_0^2}=\frac s{s+\epsilon^{\frac 1 3}} \, \frac {s^2}{s^2+\epsilon^{\frac 1 3}s+\epsilon^{\frac 2 3}}$
<br>

### Orden 1 (C-R serie):  
## $T_1(s)= \frac s {s+\frac 1 {R_1 C_1}}=\frac s {s+\epsilon^{\frac 1 3}}$  
### $\Rightarrow \, R_1 C_1=\epsilon^{-\frac 1 3} \,\rightarrow \,$ adopto: $\boxed{R_{1_N}=\epsilon^{-\frac 1 3}\approx 1,2525}\, \therefore \,\boxed{C_{1_N}=1}$  
<br>

### Orden 2 (R-C-L serie):  
## $T_2(s)= \frac {K \, s^2}{s^2+\frac {\omega_0}Q s+\omega_0^2} = \frac {s^2}{s^2+\epsilon^{\frac 1 3}s+\epsilon^{\frac 2 3}}$  
### $\Rightarrow \, K=1 \, \land \,\omega_0=\epsilon^{\frac 1 3} \, \land \,Q=1\,\,$ y los componentes:
### $\sqrt {\frac 1 {L_2 C_2}}=\omega_0=\epsilon^{\frac 1 3}\,\rightarrow \,$ adopto: $\boxed{C_{2_N=1}} \, \therefore \, \boxed{L_{2_N}=\epsilon^{-\frac 2 3}\approx 1,5686}$    
### $\frac {R_2}{L_2}=\frac {\omega_0} Q = \epsilon^{\frac 1 3} \, \Rightarrow \boxed{R_{2_N}=\epsilon^{-\frac 1 3}\approx 1,2525}$
### y de esta forma $C_{1_N} = C_{2_N} = C_N = 1\,\,$ ; $\,\,R_{1_N} = R_{2_N} = R_N = \epsilon^{-\frac 1 3}\approx 1,2525\,\,$ y $\,\, L_{2_N}=L_N=R_N^2=\epsilon^{-\frac 2 3}\approx 1,5686\,\,$ (recordar que todos estos valores son normalizados)
