### $S_{C}^{\omega_0}=\frac C{\omega_0}\frac {\partial(\omega_0)}{\partial C}= \frac C{\frac 1{R_3 C}}(\frac {-R_3}{R_3^2C}) \, \Rightarrow \, \boxed{S_{C}^{\omega_0}=-1}$

### Para obtener una transferencia pasabanda sin alterar el circuito ni componentes puedo tomar como tensi贸n de salida (que llamar茅 $V_{BP}$) la del nodo donde se encuentra la salida del amplificador operacional U1.

### Como $\,T_{BP}(s)=T_{LP}(s)\,S=-\frac {K\omega_0^2 s}{s^2+\frac {\omega_0}Q s+\omega_0^2} \,\, siendo \, \, K=H\frac{\omega_0}Q \,\,dado \, que \,\,H=\left|{T(1)}\right|$ (ganancia en frecuencia central)
### $\Rightarrow\, T_{BP}(s)=-H\frac {\omega_0}Q \frac {\omega_0^2 s}{s^2+\frac {\omega_0}Q s+\omega_0^2}$ que normalizado es $\boxed{T_{BP}(s)= -\frac {H}{Q} \frac s{s^2+\frac 1{Q}s+1}}$  
### Siguiendo con el circuito original donde $Q=3 \,y\,K=10 \Rightarrow K\,Q=\boxed{H=30}$ y as铆 $\boxed{T_{BP}(s)= -  10 \frac s{s^2+\frac 1{Q}s+1}}$  

### Simulaci贸n de funci贸n transferencia pasabanda con Python
##### Observaremos c贸mo los polos se conservan (respecto del circuito original) dado que el Q es el mismo y tambi茅n el $\omega_0$, pero aparece un cero en el origen (el otro contin煤a en el infinito). Con respecto a la respuesta de m贸dulo, el cero hace subir la pendiente desde el origen hasta $\omega_0$ en $+20 \frac {dB}{d茅c}$ y luego desde $\omega_0$ hasta el fin de la banda de paso cae con $-20 \frac {dB}{d茅c}$ (2 polos contrarrestados con un cero).


```python
w0 = 1
k = 10
q = 3

num = np.array([-k*w0**2,0]) 
den = np.array([1,w0/q,w0**2])

TS = sig.TransferFunction(num,den)

_, axes_hdl = bodePlot(TS)

plt.sca(axes_hdl[0])
plt.ylim([-20,30])

plt.gca

pzmap(TS)
GroupDelay(TS)

plt.show()
```
