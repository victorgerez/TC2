### **Hallo $\,T=\frac {V_o}{V_i}$ en funci贸n de $\omega_0$ y $Q$:**  
### $I_{R4} = I_{R4}$  
### $\frac{V_B}{R_4} = \frac {-V_o}{R_4}$  
### $V_B = -V_o$      

---------------

### $I_{R3} = I_{C}$  
### $\frac {V_A}{R_3} = -V_BSC$  
### $V_A = -R_3CSV_B \,\Rightarrow\, V_A = R_3CSV_o$  

---------------

### $I_{R1} = I_{C}+I_{R2}+I_{R3}$  
### $\frac{V_i}{R_1} = -V_A(\frac 1{R_2}+SC)-\frac{V_o}{R_3}$  
### $\frac{V_i}{R_1} = -R_3CSV_o(\frac 1{R_2}+SC)-\frac{V_o}{R_3}$  
### $\frac{V_i}{R_1} = -(\frac {R_3C}{R_2}S+R_3C^2S^2)V_o-\frac{V_o}{R_3}$  
### $\frac{V_i}{R_1} = -(R_3C^2S^2+\frac {R_3C}{R_2}S+\frac 1{R_3})V_o$  
### $T(S)=\frac{V_o}{V_i} = -\frac 1{R_1(R_3C^2S^2+\frac {R_3C}{R_2}S+\frac 1{R_3})}$  
### $\boxed{T(S)= -\frac 1{R_1R_3C^2} \, \frac 1{S^2+\frac 1{R_2C}S+\frac 1{R_3^2C^2}}}$  
### Reescribiendo $\,T(S)\,$ de la forma $\, K\frac {\omega_0^2}{S^2+\frac {\omega_0}Q S+\omega_0^2} \,\Rightarrow\, \boxed{T(S)= -\frac {R_3}{R_1} \, \frac {\frac 1{R_3^2C^2}}{S^2+\frac 1{R_2C}S+\frac 1{R_3^2C^2}}}$  
### Siendo:  
### $\boxed {K=-\frac {R_3}{R_1}}$  
### $\omega_0^2=\frac 1{R_3^2C^2}$  
### $\omega_0=\sqrt {\frac 1{R_3^2C^2}}$  
### $\boxed{\omega_0=\frac 1{R_3C}}$  
### $\frac {\omega_0}Q = \frac 1{R_2C}$  
### $\frac 1{R_3QC} = \frac 1{R_2C}$  
### $\boxed{Q=\frac {R_2}{R_3}}$  

---------------

### Normalizo en impedancia: $\mathbb Z=R_3 \,\Rightarrow \, {R_3}_N=\frac {R_3}{\mathbb Z}=1$  
### $\therefore \, K=-\frac 1{R_1'} \,;\, \omega_0=\frac 1{C'} \,;\, Q=R_2'$  
### De esta forma $\,K\,$ es gobernada por $\,R_1\,$, $\,\omega_0\,$ por $\,C\,$ y $\,Q\,$ por $\,R_2\,$ (ortogonalidad). $\,R_4\,$ puede ser elegido arbitrariamente.
### Si $\, \omega_0 = 1 \, \Rightarrow \, C'=1\, \Rightarrow \, C= \frac 1{C'}{\mathbb Z}=\frac 1{R_3}=\frac 1{1k\Omega} \, \Rightarrow \, \boxed{C=1\,mF}$  
### Si $\, Q=3 \, \Rightarrow \, R_2'=3 \, \Rightarrow \, R_2=R_2'*\mathbb Z=3R_3 \, \Rightarrow \, \boxed {R_2=3 k\Omega}$  
### Si $\, \left|{T(0)}\right|=20\,dB=10=K \, \Rightarrow \, K=\frac {1 k\Omega}{R_1}=10 \, \Rightarrow \, \boxed{R_1 = 100\Omega}$

### Para que sea una transferencia Butterworth, falta que los polos est茅n equiespaciados en 谩ngulo, por ende debe cumplirse que: $\, \alpha =\frac \pi 4 \,$  
### $\Rightarrow Q=\frac 1{2\cos (\alpha)}=\frac {\sqrt 2}2\approx0,707$  
### Si $\, Q=\frac {\sqrt 2}2 \, \Rightarrow \, R_2'=\frac {\sqrt 2}2 \, \Rightarrow \, R_2=R_2'*\mathbb Z=\frac {\sqrt 2}2R_3 \, \Rightarrow \, \boxed {R_2=\frac {\sqrt 2}2 k\Omega}$
